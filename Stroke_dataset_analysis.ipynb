```python
#Replacing the value 1 to yes in the hypertension column

df.loc[:, 'hypertension'] = df['hypertension'].replace({1: 'Yes'})
df
```

```python
#Calculating quartile ranges to be able to hande outliers in the bmi column
Q1 = df['bmi'].quantile(0.25)
Q3 = df['bmi'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
#removing outliers from bmi column
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['bmi'] >= lower_bound) & (df['bmi'] <= upper_bound)]
```

```python
#boxplot showing outliers removed fro bmi column
sns.boxplot(df['bmi'])
```

```python
#Calculating quartile ranges to be able to hande outliers in the average glucose level column
Q1 = df['avg_glucose_level'].quantile(0.25)
Q3 = df['avg_glucose_level'].quantile(0.75)
IQR = Q3 - Q1

lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
df = df[(df['avg_glucose_level'] >= lower_bound) & (df['avg_glucose_level'] <= upper_bound)]
```

```python
#boxplot showing outliers removed fro avg_glucose_level column
sns.boxplot(df['avg_glucose_level'])
```

```python
#Explore the distribution of age, avg_glucose_level, and bmi

# Set up the figure with subplots
fig, axes = plt.subplots(nrows=3, ncols=1, figsize=(10, 12))

# Distribution plots for age, avg_glucose_level, and bmi
sns.histplot(df['age'], kde=True, ax=axes[0]).set_title('Distribution of Age')
sns.histplot(df['avg_glucose_level'], kde=True, ax=axes[1]).set_title('Distribution of Avg Glucose Level')
sns.histplot(df['bmi'], kde=True, ax=axes[2]).set_title('Distribution of BMI')

plt.tight_layout()
plt.show()
```

```python
#Visualize the distribution of strokes in the dataset:

# Countplot for strokes
df['stroke'].value_counts().plot(kind='pie',autopct = '%1.1f%%')
plt.title('Strokes distribution ')
```

```python
#Explore the correlation between hypertension, heart_disease, and strokes:

# Crosstab to analyze correlation
correlation_table = pd.crosstab(index=df['hypertension'], columns=[df['heart_disease'], df['stroke']], margins=True, margins_name="Total")
print(correlation_table)


# Plotting the heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(correlation_table, annot=True, cmap='coolwarm', fmt='g', cbar=True)
plt.title('Correlation Heatmap: Hypertension, Heart Disease, and Strokes')
plt.show()
#This heatmap will visually represent the correlations between hypertension, heart disease, and strokes. 
#Positive correlations will be represented by warmer colors, while negative correlations will be represented 
#by cooler colors. The annotations will display the actual count values.
```
